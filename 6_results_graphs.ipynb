# **Results Graphs**

# Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.patches as mpatches
```

# Percent Error

```python
# Uniform Distribution 1
unif1 = {'sample_size': [10000, 100000, 1000000, 3000000],
         'error_recon': [15, 35, 4, 0.4],
         'error_chao1': [22, 15, 1, 0.08],
         'error_ace': [15, 35, 4, 0.4],
         'error_powerlaw': [88, 36, 0.4, 1]}
unif1['sample_percent'] = [100*s/5037134 for s in unif1['sample_size']]
unif1 = pd.DataFrame(data=unif1)

# Uniform Distribution 2
unif2 = {'sample_size': [10000, 100000, 1000000, 3000000, 10000000, 20000000, 30000000, 40000000],
         'error_recon': [59, 17, 52, 10, 4, 2, 0.9, 0.2],
         'error_chao1': [33, 22, 15, 7, 1, 0.14, 0.08, 0.1],
         'error_ace': [32, 22, 15, 8, 2, 0.5, 0.1, 0.07],
         'error_powerlaw': [99, 88, 37, 10, 1, 1, 1, 1]}
unif2['sample_percent'] = [100*s/50485039 for s in unif2['sample_size']]
unif2 = pd.DataFrame(data=unif2)

# Mixed Poisson Distribution 1
mp1 = {'sample_size': [10000, 100000, 300000],
         'error_recon': [67, 14, 6],
         'error_chao1': [11, 6, 0.5], 'error_ace': [11, 9, 2],
         'error_powerlaw': [88, 24, 9]}
mp1['sample_percent'] = [100*s/364902 for s in mp1['sample_size']]
mp1 = pd.DataFrame(data=mp1)

# Mixed Poisson Distribution 2
mp2 = {'sample_size': [10000, 100000, 300000, 1000000, 2000000],
         'error_recon': [59, 68, 20, 14, 8],
         'error_chao1': [8, 13, 12, 6, 2], 'error_ace': [10, 14, 13, 9, 5],
         'error_powerlaw': [99, 88, 69, 24, 1]}
mp2['sample_percent'] = [100*s/3650542 for s in mp2['sample_size']]
mp2 = pd.DataFrame(data=mp2)
```

```python
unif1, unif2, mp1, mp2
```

```python
fig, axs = plt.subplots(2, 2, figsize = (10, 8), constrained_layout = True) # sharex=True, sharey=True

# Unif1
axs[0, 0].set_title('Uniform 100,000')
axs[0, 0].set_title('a)', loc='left', fontsize='medium')
axs[0, 0].plot(unif1['sample_percent'], unif1['error_recon'], 'tab:orange')
axs[0, 0].plot(unif1['sample_percent'], unif1['error_chao1'], 'tab:blue')
axs[0, 0].plot(unif1['sample_percent'], unif1['error_ace'], 'tab:green')
axs[0, 0].plot(unif1['sample_percent'], unif1['error_powerlaw'], 'tab:purple')
axs[0, 0].set(ylabel='Percent Error')
axs[0, 0].set(xlabel='Percent Sampled')
axs[0, 0].set_xlim(0, 59)
axs[0, 0].set_ylim(0, 100)

# MixedPoisson1
axs[0, 1].plot(mp1['sample_percent'], mp1['error_recon'], 'tab:orange')
axs[0, 1].plot(mp1['sample_percent'], mp1['error_chao1'], 'tab:blue')
axs[0, 1].plot(mp1['sample_percent'], mp1['error_ace'], 'tab:green')
axs[0, 1].plot(mp1['sample_percent'], mp1['error_powerlaw'], 'tab:purple')
axs[0, 1].sharey(axs[0, 0])
axs[0, 1].set_title('Mixed Poisson 100,000')
axs[0, 1].set_title('b)', loc='left', fontsize='medium')
axs[0, 1].set(ylabel='Percent Error')
axs[0, 1].set(xlabel='Percent Sampled')
axs[0, 1].set_xlim(0, 82)
axs[0, 1].set_ylim(0, 100)

# Row 2:
# Unif2
axs[1, 0].plot(unif2['sample_percent'], unif2['error_recon'], 'tab:orange')
axs[1, 0].plot(unif2['sample_percent'], unif2['error_chao1'], 'tab:blue')
axs[1, 0].plot(unif2['sample_percent'], unif2['error_ace'], 'tab:green')
axs[1, 0].plot(unif2['sample_percent'], unif2['error_powerlaw'], 'tab:purple')
axs[1, 0].set(ylabel='Percent Error')
axs[1, 0].set(xlabel='Percent Sampled')
axs[1, 0].set_title('Uniform 1,000,000')
axs[1, 0].set_title('c)', loc='left', fontsize='medium')
axs[1, 0].set_xlim(0, 79)
axs[1, 0].set_ylim(0, 100)

axs[1, 1].plot(mp2['sample_percent'], mp2['error_recon'], 'tab:orange')
axs[1, 1].plot(mp2['sample_percent'], mp2['error_chao1'], 'tab:blue')
axs[1, 1].plot(mp2['sample_percent'], mp2['error_ace'], 'tab:green')
axs[1, 1].plot(mp2['sample_percent'], mp2['error_powerlaw'], 'tab:purple')
axs[1, 1].set_title('Mixed Poisson 1,000,000')
axs[1, 1].set(ylabel='Percent Error')
axs[1, 1].set(xlabel='Percent Sampled')
axs[1, 1].set_title('d)', loc='left', fontsize='medium')
axs[1, 1].set_xlim(0, 54)
axs[1, 1].set_ylim(0, 100)

pop_a = mpatches.Patch(color='tab:orange', label='Recon')
pop_b = mpatches.Patch(color='tab:blue', label='Chao1')
pop_c = mpatches.Patch(color='tab:green', label='ACE')
pop_d = mpatches.Patch(color='tab:purple', label='Power Law')
fig.legend(handles=[pop_a, pop_b, pop_c, pop_d], bbox_to_anchor=(1.15, 1.0)) #x, y
plt.subplots_adjust(right=0.9)

plt.savefig('results_percent_error.png', dpi=300, bbox_inches='tight')
```
